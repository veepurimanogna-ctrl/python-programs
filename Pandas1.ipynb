```python
import pandas as pd

s = pd.Series([10,20,30,40])
print(s)
```

```python
student = {
    "Manogna":85,
    "Manasa":92,
    "Shiva":78
}
s = pd.Series(student)
print(s)
```

```python
print(s[0])
print(s["Manogna"])
```

```python
print(s.index)
print(s.values)
print(s.dtype)
print(s.shape)
print(s.size)
```

```python
marks = [80,90,75,95]

s = pd.Series(marks)

print(s)
```

```python
print(s.index)
print(s.values)
print(s.dtype)
print(s.shape)
print(s.size)
```

```python
print("Sum:", s.sum())
print("Mean:", s.mean())
print("Max:", s.max())
print("Min:", s.min())
print("Count:", s.count())
print("Median:", s.median())
print("Standard Deviation:", s.std())
```
